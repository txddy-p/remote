```python
#Question 1
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np

t = np.linspace(0, 10, 200);
x = np.cos(4*t)
y = np.sin(4*t)

fig, (Re, Im) = plt.subplots(2,1)
fig.suptitle('Question 1')
Re.plot(t, x);
Re.set_ylabel('Re');Re.set_xlabel('t');

Im.plot(t, y);
Im.set_ylabel('Im');Im.set_xlabel('t');
fig.show()
```

```python
#Question 3
import sympy as sp
v,g,t = sp.symbols('v g t')
f = v*t -(0.5*g*(t**2))
exp = f.subs([(g,9.8),(v,20)])

sp.plot(exp, (t,0,10), xlabel="t [s]", ylabel="s(t) [m]", title="Height vs time");
```

```python
#Question 4
t = sp.symbols('t')
acc = 0.3+sp.cos(t)

vt = sp.integrate(acc)-0.2 #velocity is the integral of acceleration
sp.plot(vt, (t,0,5), xlabel="t [s]", ylabel="v(t) [m/s]", title="Velocity vs time")

st = sp.integrate(vt)+1.1 #displacement is the integral of velocity
sp.plot(st, (t,0,5), xlabel="t [s]", ylabel="s(t) [m]", title="Distance vs time");
```
